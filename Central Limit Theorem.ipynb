# Central Limit Theorem

```python
def draw_sample_statistics(data, func, n_samples = 1000, sample_size = 10):
    # Initialize array of sample statistics: sample_stats
    sample_stats = np.empty(n_samples)
    # Generate sample_stats
    for i in range(n_samples):
        sample_stats[i] = func(np.random.choice(data, sample_size))
    return sample_stats
```

```python
def central_limit_theorem(data, n_samples = 1000, sample_size = 500, min_value = 0, max_value = 1338):
    """ Use this function to demonstrate Central Limit Theorem. 
        data = 1D array, or a pd.Series
        n_samples = number of samples to be created
        sample_size = size of the individual sample"""
    
    sample_means = draw_sample_statistics(data, np.mean, n_samples, sample_size)
    
    plt.figure(figsize=(15,5))

    plt.subplot(1,2,1)
    sns.distplot(sample_means)
    plt.title("Sampling Distribution. \n \u03bc = {} & Std Error = {}".format(round(sample_means.mean(), 3), 
                                                                              round(sample_means.std(), 3)))
    plt.xlabel('data')
    plt.ylabel('freq')

    plt.subplot(1,2,2)
    sns.distplot(data)
    plt.title("population Distribution. \n \u03bc = {} & \u03C3 = {}".format(round(data.mean(), 3),
                                                                             round(data.std(), 3)))
    plt.xlabel('data')
    plt.ylabel('freq')

    plt.show()
```

```python
central_limit_theorem(series1, n_samples = 5000, sample_size = 500)
```
