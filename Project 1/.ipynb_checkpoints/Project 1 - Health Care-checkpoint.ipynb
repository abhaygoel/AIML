## 4a. Do charges of people who smoke differ significantly from the people who don't?

(Null Hypothesis) $H_o$ - Charges of people who smoke are the same as that of people who do not smoke.

$\mu_0 - \mu_1 = 0$

(Alternative Hypothesis) $H_a$ - Charges are different for the two categories of people

$\mu_0 - \mu_1 \neq 0$

- This is a two tailed test because difference may lie in either direction
- Use two sample t-test

```python
''' Draw a Box plot to get a feel for charges per the category of people '''
sns.boxplot(y=insurance['charges'],x=insurance['smoker'])
```

```python
''' Separate data for smoker and non-smoker '''
smoker = insurance['smoker']=='yes'
smoker_charges = insurance[smoker].charges 

non_smoker = insurance['smoker']=='no'
non_smoker_charges = insurance[non_smoker].charges

# two-sample t-test
t_statistic, p_value = ttest_ind(smoker_charges, non_smoker_charges)
if p_value < 0.05:
    print(f'With a p-value of {round(p_value,4)} the difference is significant aka We reject the null hypothesis')
    print('Charges for smokers are NOT the same from that of non-smokers.')
else:
    print(f'With a p-value of {round(p_value,4)} the difference is not significant. aka We fail to reject the null|')
    print('Charges for smokers are significantly same from that of non-smokers')
```

## 4b. Does bmi of males differ significantly from that of females?

(Null Hypothesis) $H_o$ - BMI of Males is the same as that of Females

$\mu_0 - \mu_1 = 0$

(Alternative Hypothesis) $H_a$ - BMI of Males differes significantly from that of Females

$\mu_0 - \mu_1 \neq 0$

- This is a two tailed test because the difference may lie in either direction
- Use two sample t-test

```python
''' Draw a Box plot to get a feel for bmi per the sex of people '''
sns.boxplot(y=insurance['bmi'],x=insurance['sex'])
```

```python
''' Separate data for Male and Female '''
male = insurance['sex']=='male'
male_bmi = insurance[male].bmi 

female = insurance['sex']=='female'
female_bmi = insurance[female].bmi

# two-sample t-test
t_statistic, p_value = ttest_ind(male_bmi, female_bmi)

if p_value < 0.05:
    print(f'With a p-value of {round(p_value,4)} the difference is significant aka We reject the null hypothesis.')
    print('BMI is NOT the same between Male and Female.')
else:
    print(f'With a p-value of {round(p_value,4)} the difference is not significant aka We FAIL to reject the null hypothesis.')
    print('BMI is the same between Male and Female.')

```

## 4d. Is the distribution of bmi across women with no children, one child and two children, the same ?

(Null Hypothesis) $H_o$ - No difference between bmi

(Alternative Hypothesis) $H_a$ - Some difference between bmi exists somewhere

- Use ANOVA
- Assumptions:
  - Samples are independent
  - Each sample is from a normally distributed population.
  - The population standard deviations of the groups are all equal. This property is known as homoscedasticity.
  

```python
# group data for female based on number of children
female = insurance['sex']=='female'
children_0 = insurance['children']==0
children_1 = insurance['children']==1
children_2 = insurance['children']==2

bmi_0 = insurance[female & children_0].bmi.values
bmi_1 = insurance[female & children_1].bmi.values
bmi_2 = insurance[female & children_2].bmi.values
#print(bmi_0.count(),bmi_1.count(),bmi_2.count() )

# check homogeneity of variance
s, pval = levene(bmi_0,bmi_1,bmi_2)
if (pval > 0.05):
    #print(f'Assumption: with a levene p_value of {round(pval,4)}, the variances are homogeneous. We can proceed with ANOVA')
    #pass data to anova
    fstat, p_value = f_oneway(bmi_0,bmi_1,bmi_2)
    if pval < 0.05:
        print(f'With a p-value of {round(p_value,4)} the difference is significant aka We reject the null hypothesis')
    else:
        print(f'With a p-value of {round(p_value,4)} the difference is not significant aka We fail to reject the null hypothesis.')

else:
    print(f'with a levene p_value of {round(pval,4)}, the variances are not homogeneous. We can not proceed with ANOVA')

```
