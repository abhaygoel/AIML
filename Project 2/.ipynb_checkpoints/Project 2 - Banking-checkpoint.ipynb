#### Data Cleanup
- Experience column contains negative values.
- Apply the following logic to clean up -ve values :
    - For the record with the ID, get the value of Age column
    - For the record with the ID, get the value of Education column
    - Filter the records matching the above criteria from the data frame which has records with positive experience and take the median
    - Apply the median back to the location which had negative experience

```python
# There are 52 rows with -ve experience
banking[banking['Experience'] < 0]['Experience'].count()

#clean the negative variable
dfExp = banking.loc[banking['Experience'] >0]
negExp = banking.Experience < 0
column_name = 'Experience'
# getting the customer ID with negative experience
mylist = banking.loc[negExp]['ID'].tolist()
plist = banking.loc[negExp]['Personal Loan'].sum()

for id in mylist:
    age = banking.loc[np.where(banking['ID']==id)]["Age"].tolist()[0]
    education = banking.loc[np.where(banking['ID']==id)]["Education"].tolist()[0]
    df_filtered = dfExp[(dfExp.Age == age) & (dfExp.Education == education)]
    exp = df_filtered['Experience'].median()
    banking.loc[banking.loc[np.where(banking['ID']==id)].index, 'Experience'] = exp
    
    
# checking if there are records with negative experience
banking[banking['Experience'] < 0]['Experience'].count()

```

```python
# There are scenarios where required Age and Education combination does not exist. Removing these rows.
print(banking.isnull().sum())
banking.dropna(inplace=True)
print(banking.isnull().sum())
```

#### Applying Models
- Split data into train and test

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
from sklearn import metrics

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# 30% of the data will be used for testing
test_size= 0.30
seed = 100


# X and y features
y1 = banking['Personal Loan']
X1 = banking.drop(['ID', 'Personal Loan', 'Experience', 'Online', 'CreditCard', 'ZIP Code', 'Age'], axis =1)

x_train1, x_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=test_size, random_state=seed)

```

### Logistic Regression

```python
# Logistic Regression

LR = LogisticRegression(solver='lbfgs', max_iter=1000)
# fit model

LR.fit(x_train1, y_train1)
```

#### Get the Coefficient for each feature.


```python
name = X1.columns
coef = LR.coef_[0]

print(pd.DataFrame([name,coef],index = ['Name','Coef']).transpose())
```

#### Observation
- CD Account has the largest coeff followed by Education.

### Use the Model to predict on x_test

```python
# predict with model
y_pred = LR.predict(x_test1)
```

```python
metrics.accuracy_score(y_test1,y_pred)
```

```python
banking['Personal Loan'].value_counts()
```

#### Accuracy Score
Accuracy score is the percentage of correct predictions out of all predictions made.
- Thats a pretty good score - almost 95% 
- In our entire dataset there are 4491 "No" and 480 "Yes" and a total of 4971 observations. 
- __If we guessed that nobody was going to convert and therefore 'No' every time, we would be correct 89% of the time (according to our data). That being said, this accuracy is better than baseline.__

#### Confusion Matrix

```python
cm = metrics.confusion_matrix(y_test1,y_pred)
print(cm)

class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
# annot_kws = {"ha": 'center',"va": 'top'}
sns.heatmap(pd.DataFrame(cm), annot=True, cmap="YlGnBu" ,fmt='d')
ax.xaxis.set_label_position("top")
#plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

#### Observation
- Out of 1352 No it predicted 1328 correctly (True Negative)
- Out of 140 instances of Yes it correctly predicted 86.

A problem revealed by this confusion matrix is that the dataset is highly unbalanced, with nearly all client actually do not have a Personal Loan. This infers that the accuracy score is biased, and further evaluation should be carried out to determine the accuracy of logistic regression model.

#### ROC AUC

```python
# Probability estimates
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score


# Keep Probabilities for the positive Outcomes only
y_pred_proba = LR.predict_proba(x_test1)[:,1]

auc = roc_auc_score(y_test1, y_pred_proba)
fpr, tpr, _ = roc_curve(y_test1, y_pred_proba)
plt.plot(fpr, tpr, marker='.', label="auc="+str(auc))
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# show the legend
plt.legend()
# show the plot
plt.show()
```

__Model predicts the majority class (class 0) in all cases on the test set. The score is very misleading.__

#### Precision-Recall Curve

```python
from sklearn.metrics import auc

precision, recall, _ = precision_recall_curve(y_test1, y_pred_proba)

f1= f1_score(y_test1, y_pred)

auc = auc(recall, precision)
plt.plot(recall, precision, marker='.', label="auc="+str(auc))
# axis labels
plt.xlabel('Recall')
plt.ylabel('Precision')
# show the legend
plt.legend()
# show the plot
plt.show()
print('f1=%.3f auc=%.3f' % (f1, auc))
```

We can see that the model is penalized for predicting the majority class in all cases. The scores show that the model that looked good according to the ROC Curve is in fact barely skillful when considered using using precision and recall that focus on the positive class.

#### When to Use ROC vs. Precision-Recall Curves?

- ROC curves should be used when there are roughly equal numbers of observations for each class.
- Precision-Recall curves should be used when there is a moderate to large class imbalance.

The reason for this recommendation is that ROC curves present an optimistic picture of the model on datasets with a class imbalance.

### Compare classification algorithms
Three different classification algorithms (Logistic Regression, K-Neighbors Classifier, and Gaussian NB) are run on the dataset and identify the best performing one.

```python
models = []
models.append(('LR', LogisticRegression(solver='lbfgs', max_iter=1000)))
models.append(('KNN', KNeighborsClassifier(n_neighbors= 21 , weights = 'uniform', metric='euclidean')))
models.append(('NB', GaussianNB()))
```

```python
results_c = []
names_c = []

for name, model in models:
    # define how to split off validation data ('kfold' how many folds)
    kfold = KFold(n_splits=10, random_state=seed)    
    # train the model
    cv_results = cross_val_score(model, x_train1, y_train1, cv=kfold, scoring='accuracy', error_score='raise')    
    results_c.append(cv_results)
    names_c.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

#### Logistic regression is the best performing model.


```python
# Plot results
fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111)
plt.boxplot(results_c)

fig.suptitle('Algorithm Comparison')
ax.set_xticklabels(names_c)
plt.show()
```


Among all algorithms, logistic regression achieved an accuracy of about 95%, suggesting a high level of strength of this model to classify the customer response given all the defined customer features.
